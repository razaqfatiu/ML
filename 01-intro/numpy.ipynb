Cross Product 

```python
u = np.array([1, 20, 30])
v = np.array([1, 2, 3])
```

```python
def vector_vector_multip(u, v):
    assert u.shape[0] == v.shape[0]
    result = 0.0
    for i in range(u.shape[0]):
        result += (u[i] * v[i])
    return result
        
```

```python
vector_vector_multip(u, v)
```

```python
u.dot(v)
```
