Cross Product 

```python
u = np.array([1, 20, 30])
v = np.array([1, 2, 3])
```

```python
def vector_vector_multip(u, v):
    assert u.shape[0] == v.shape[0]
    result = 0.0
    for i in range(u.shape[0]):
        result += (u[i] * v[i])
    return result
```

```python
vector_vector_multip(u, v)
```

```python
v = np.array([2,4,5,6])

U = np.array([
    [2,4,5,6],
    [1,2,1,2],
    [3,1,2,1]
])
```

```python
U.shape
```

```python
def vector_by_matrix(U, v):
    assert U.shape[1] == v.shape[0]

    rows = U.shape[0]

    result = np.zeros(rows)

    for i in range(rows):
        result[i] = vector_vector_multip(U[i], v)

    return result
```

```python
vector_by_matrix(U, v)
```

```python
def matrix_by_matrix (U, V):
    assert U.shape[1] == V.shape[0]

    rows = U.shape[0]
    cols = V.shape[1]

    result = np.zeros((rows, cols))
    
    for i in range(rows):
        vi = V[:, i]
        result[:, i] = vector_by_matrix(U , vi)
        print(result)
    return result
    
```
